```python
age_file = open('Age_tagged_corpora.csv')
dep = []
other = []
for i in age_file:
    list_age = i.strip().split(',')
    if list_age[0][:3]=='dep':
        dep.append([int(list_age[0][11:]),list_age[1]])
    elif list_age[0][:3] == 'oth':
        other.append([int(list_age[0][13:]),list_age[1]])
```

```python
dep.sort(key=lambda x: x[0])
other.sort(key=lambda x: x[0])
```

```python
age_feature = []
for i in dep:
    if i[1] == 'Under':
        age_feature.append(0)
    else:
        age_feature.append(0)
```
